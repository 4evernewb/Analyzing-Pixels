```python
from PIL import Image
from IPython.display import display
import inspect 
file1 = "nemo1.png"
image1 = Image.open(file1)
print("This type of image is "+str(type(image1)))
inspect.getmro(type(image1))
w,h = image1.size
print(w,h)
print(image1.mode)
print(image1)
image1.show()
display(image1)
```

```python
#Analyzing the first 36 pixels coordinate wise.
import pandas as pd
from pandas import ExcelWriter

writer = pd.ExcelWriter('nemo_coloring.xlsx')


colors = []
for i in range(1,10):
    for j in range(1,5):
        colors1 = []
        coloring = image1.getpixel((i,j))
        colors1.append(i)
        colors1.append(j)
        colors1.append(coloring[0])
        colors1.append(coloring[1])
        colors1.append(coloring[2])
        colors1 = tuple(colors1)
        colors.append(colors1)
        
df = pd.DataFrame(colors,columns=['X Pixel','Y Pixel','Red(R)','Green(G)','Blue(B)'])
print(df)
df.to_excel(writer,sheet_name="nemo1",index=False)
workbook = writer.book
worksheet = writer.sheets["nemo1"]
cell_format1 = workbook.add_format()
cell_format1.set_font_color('red')
worksheet.set_column('C:C', None, cell_format1)

cell_format2 = workbook.add_format()
cell_format2.set_font_color('green')
worksheet.set_column('D:D', None, cell_format2)

cell_format3 = workbook.add_format()
cell_format3.set_font_color('blue')
worksheet.set_column('E:E', None, cell_format3)

writer.close()
```
